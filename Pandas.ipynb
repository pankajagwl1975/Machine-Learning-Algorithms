#### pandas.Series : Pandas Series is a One Dimensional indexed array. It is most similar to the NumPy array. pandas.Series is a method to create a series.

```python
import pandas as pd
```

```python
pd.__version__
```

##### Series is similar to python list but series have additional functionality, methods, and operators, because of these series is advanced than a list.
##### pandas.core.series.Series means series is a one-dimensional array, which can store indexed data

###### Methods of Creating a Series

###### Creating series from list

```python
list_1 = [1, 2, -3, 4.5, 'indian']
print(list_1)
series1 = pd.Series(list_1)
print(series1)
```

```python
type(series1)
```

###### Creating empty series : Empty series is like an empty list, we can create empty series using an empty list

```python
empty_s = pd.Series([])
print(empty_s)
```

###### Creating Series using a different method : List inside the series

```python
series2 = pd.Series([1,2,3,4,5])
print(series2)
```

###### In index parameter, default index is start from 0 to n (0,1,2,….n) when index is not identified. Here we are creating series with the index parameter. Index length should have equal to the number of data values, otherwise, it shows error.

```python
series2 = pd.Series([1,2,3,4,5], index = ['a', 'b', 'c', 'd', 'e'])
print(series2)
```

###### We can change index to any numbers, alphabates, names etc.
###### Changing data type of series
###### Above you can see dtype: int64, this means our data type has stored in integer 64 bit. we can change the data type of series

```python
series2 = pd.Series([1,2,3,4,5], index = ['a', 'b', 'c', 'd', 'e'], dtype = float)
print(series2)
```

###### Creating series from scalar values : scalar values means single value. e.g. 1, 0.5, ‘indian’

```python
s3_scalar = pd.Series(2)
print(s3_scalar)
```

###### for more data values index should be needed.

###### Creating series from python dictionary

```python
s4_dict = pd.Series({'a':1, 'b':2, 'c':3})
print(s4_dict)
```

###### Pandas Join() Method

```python
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]})
 
 
df2 = pd.DataFrame({'C': [4,5,6],
                   'D': [40,50,60]})
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df1.join(df2, how = 'right')
```

```python
df1.join(df2, how = 'inner')
```

```python
df1.join(df2, how = 'outer')
```

```python
df1 = pd.DataFrame({'A': [1,2,3],
                   'B': [10,20,30]})
 
df2 = pd.DataFrame({'A': [4,5,6],
                   'D': [40,50,60]})
```

```python
df1.join(df2, lsuffix = '_1')
```

```python
df1.join(df2, rsuffix = '_1')
```
