# Data Visualiztion
### Data visualization allows us to quickly interpret the data and adjust different variables to see their effect. 
### It is easier to understand the result by charts and graphs rather than reading tables or thousands of line of raw data. 
### Our brain can understand visual things better and it helps to take the decisions.

### Purpose:
- Observe the patterns
- Identify extreme values that could be anomalies
- Easy interpretation

### Matlplotlib is a Python module for visualization. 
#### Matplotlib allows you to easily make line graphs, pie chart, histogram and other professional grade figures.
#### Using Matplotlib you can customize every aspect of a figure. When used within IPython, Matplotlib has interactive features like zooming and panning. 
#### It supports different GUI back ends on all operating systems, and can also export graphics to common vector and graphics formats: PDF, SVG, JPG, PNG, BMP, GIF, etc.

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.plot([1,2,3],[4,2,7])
plt.show()
```

#### We can also change the colour, type ('-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'), width of the line.
#### Markrs can also be added on each data points

```python
#define colour, line style and width
plt.plot([1,2,3],[4,2,7], color ='red', ls = '--', lw = 10)
plt.show()
```

```python
#Defining markers and marker width
plt.plot([1,2,3],[4,2,7],marker = '*', mew = 10)
```

```python
x1=[1,2,3]
y1=[2,4,2]
x2=[3,4,5]
y2=[6,8,6]
plt.plot(x1,y1,label='first plot',color='red')
plt.plot(x2,y2,label='second plot',color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.title('Two Plots in Same Graph')
plt.show()
```

### Plot line chart from Pandas dataframe
#### To plot line chart we can also use xticks and yticks are used to put significant range on axis.

```python
sales_Delhi = [34, 45, 33, 45, 49]
sales_Pune = [12, 13, 6, 14, 10]
sales_Mumbai = [67, 78, 90, 75, 85]
colors = ['red','green', 'black']
sales = pd. DataFrame({'Delhi': sales_Delhi, 'Pune': sales_Pune, 'Mumbai': sales_Mumbai})
sales.plot(xticks = range(0,5), yticks = range(0,100,20), color = colors)
```

### Scatter Plot
#### A scatter plot is a set of points that represents the values obtained for two different variables plotted on a horizontal and vertical axes
#### When to use Scatter Plot?
- Scatter plots are used to convey the relationship between two numerical variables. (like traffic intensity and number of cars, number of research paper published and years of experience.) 
- Scatter plots are sometimes called correlation plots because they show how two variables are correlated

```python
x=[1,2,3,4,5,6,7,8]
y=[4,3,5,3,2,3,4,9]
plt.scatter(x,y,label='Scatter',marker="*",s=25)
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

### Bar plot 
#### A bar plot is a plot that presents categorical data with rectangular bars with lengths proportional to the counts that they represent.
#### When to use bar plot? 
#### To represent the frequency distribution of categorical variables 
#### A bar diagram makes it easy to compare sets of data between different groups

```python
# to plot bar chart with dataframe object plot function will be used.
sales.plot(kind = 'bar')
```

```python
x=[1,2,4,6,8]
y=[8,5,3,7,4]
plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Chart')
plt.show()
```

```python
x1=[1,2,4,6,8]
y1=[8,5,3,7,4]
x2=[3,6,7,9,13]
y2=[4,2,8,7,5]
plt.bar(x1,y1,label='first plot',color='red')
plt.bar(x2,y2,label='second plot',color='magenta')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Chart')
plt.show()
```

```python
count=[900,100,10]
fueltype=('Petrol','Diesel','CNG')
index=range(0,3)
plt.bar(index,count,color=['red','green','cyan'],edgecolor='black')
plt.title('Bar Chart of fuel type')
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.show()
```

### Histogram 
#### It is a graphical representation of data using bars of different heights 
#### Histogram groups numbers into ranges and the height of each bar depicts the frequency of each range or bin
#### When to use histograms? 
#### To represent the frequency distribution of numerical variables

```python
plt.hist(df1['TotalBill'],color='red',edgecolor='black',bins=5)
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.show()
```

### Pie Chart
#### Pie chart represent whole data as a circle. 
#### Different categories makes slices along the circle based on their properties.
#### Pie chart is useful if we want to compare each category in a common scale.

```python
a = [3,4,5,7,12]
plt.pie(a,labels=['AA','BB','CC','DD','EE'])
```

```python
#We can define colors for each categories
color_list = ['Red','Green','Blue','Yellow','Grey']
plt.pie(a,labels=['AA','BB','CC','DD','EE'],colors=color_list)
plt.show()
```

```python
slices=[95,34,30,25,12]
channels=['sony','ZEE','colors','etv','star plus']
cols=['red','blue','green','cyan','y']
plt.pie(slices,labels=channels,colors=cols,shadow=True,autopct = '%2.2f%%')
plt.show()
```

### Box Plot
#### Bos Plot is used to understand the variable spread. In a box plot, rectangle top boundary represents third quantile, bottom boundary represents first quantile and line in the box indicates median. Vertical line at the top indicate the maximum value and vertical line at the bottom indicates the minimum value.

```python
plt.boxplot(df1['Size'])
plt.show()
```

### Heatmap
#### Heatmap is a graphical representation of 2D dataset. Each data value represent in a matrix which can be shown in colour form.
#### Heatmap can be used in feature engineering when correlation can be found.

```python
df3=pd.read_csv("C:\\Users\\Sapna\\Desktop\\ML Classes\\visualization\\head_brain1.csv")
df3
```

```python
sns.heatmap(df3)
```

```python
plt.figure(figsize=(30,20))
sns.heatmap(df3, vmin = 0, vmax = 4500, cmap ='coolwarm', annot=True)
```

```python
plt.figure(figsize=(30,20))
ant_ar={"fontsize":10, "fontstyle":"italic", "color":'k',"alpha":.9,'rotation':'vertical','backgroundcolor':'w'}
sns.heatmap(df3, vmin = 0, vmax = 4500, cmap ='coolwarm', annot=True,annot_kws=ant_ar, linewidth = 4, linecolor='k', cbar=False)
```

```python
plt.figure(figsize=(40,40))
ant_ar={"fontsize":10, "fontstyle":"italic", "color":'k',"alpha":.9,'rotation':'vertical','backgroundcolor':'w'}
cbk = {'orientation':'horizontal', 'shrink':1,'extend':'min', 'drawedges':True}
sns.heatmap(df3, vmin = 0, vmax = 4500, cmap ='coolwarm', annot=True,annot_kws=ant_ar, linewidth = 4, linecolor='k', cbar=True, cbar_kws=cbk)
```
